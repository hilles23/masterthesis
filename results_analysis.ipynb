```python
import pandas as pd
```

```python
#load the data

results_age = pd.read_csv('/home/hilles/answers_bbq_age.csv')
results_dis = pd.read_csv('/home/hilles/answers_bbq_disability_status.csv')
results_gen = pd.read_csv('/home/hilles/answers_bbq_gender_identity.csv')
results_nat = pd.read_csv('/home/hilles/answers_bbq_nationality.csv')
results_app = pd.read_csv('/home/hilles/answers_bbq_physical_appearance.csv')
results_rac = pd.read_csv('/home/hilles/answers_bbq_race_ethnicity.csv')
results_rel = pd.read_csv('/home/hilles/answers_bbq_religion.csv')
results_ses = pd.read_csv('/home/hilles/answers_bbq_ses.csv')
results_sor = pd.read_csv('/home/hilles/answers_bbq_sexual_orientation.csv')
```

```python
# print the shape to see if it matches the numbers from the preprocessing part (bbq.groupby('category').count())

print("results_age",results_age.shape) #correct
print("results_dis",results_dis.shape) #1546 instead of 1548, 2 ambiguous questions got "lost" somehow...
print("results_gen",results_gen.shape) #correct
print("results_nat",results_nat.shape) #correct
print("results_app",results_app.shape) #correct
print("results_rac",results_rac.shape) #correct
print("results_rel",results_rel.shape) #correct
print("results_ses",results_ses.shape) #6848 instead of 6856, but still enough questions 
print("results_sor",results_sor.shape) #correct
```
